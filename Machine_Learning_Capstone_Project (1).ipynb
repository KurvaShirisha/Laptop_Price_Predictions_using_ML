### **MODEL TRAING**

```python
df.head()
```

```python
# Nominal encoding to these columns.Becuase these columns does not have any order
df = pd.get_dummies(df, columns=['Company', 'TypeName', 'Cpu_Brand','Gpu_Brand','os'],drop_first=True)
```

```python
df.head()
```

```python
# separeting the independent variables and dependent variable
X = df.drop(columns=['Price'])
y = np.log(df['Price'])
```

```python
X.head()
```

***Feacture*** ***scaling***

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
scaled=ss.fit_transform(X)
```

```python
X_new=pd.DataFrame(scaled,columns=X.columns)
```

```python
X_new.head()
```

### **Linear Regression**

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)

# Train machine learning model
from sklearn.linear_model import LinearRegression

# Initialize model
lr_model = LinearRegression()

# Train model
lr_model.fit(X_train, y_train)

```

```python
#Predicting the X_test
ycap=lr_model.predict(X_test)
```

```python
# Evaluting the model using R2 score,mean square error
from sklearn import metrics
R2_score=metrics.r2_score(y_test,ycap)
print("R2_score: ",R2_score)
mse=np.sqrt(metrics.mean_squared_error(y_test,ycap))
print("Mean square error: ",mse)
```

### **Random** **Forest** **Regressior**

```python
# Train machine learning model
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()

rf_model.fit(X_train, y_train)
```

```python
# Predicting X_test
ycap=rf_model.predict(X_test)
```

```python
# Evalution of the model
from sklearn import metrics
R2_score=metrics.r2_score(y_test,ycap)
print("R2_score: ",R2_score)
mse=np.sqrt(metrics.mean_squared_error(y_test,ycap))
print("Mean square error: ",mse)
```

### **GradientBoosting Regressor**

```python
# Train machine learning model
from sklearn.ensemble import GradientBoostingRegressor
gb_model = GradientBoostingRegressor()

gb_model.fit(X_train, y_train)
```

```python
# Predicting the model
ycap=gb_model.predict(X_test)
```

```python
# Evalution of the model
from sklearn import metrics
R2_score=metrics.r2_score(y_test,ycap)
print("R2_score: ",R2_score)
mse=np.sqrt(metrics.mean_squared_error(y_test,ycap))
print("Mean square error: ",mse)
```
